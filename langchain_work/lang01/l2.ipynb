### Models Integration with OpenAI, Google Gemini and Groq

```python
import os
from dotenv import load_dotenv
load_dotenv()


os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY")
os.environ["GOOGLE_API_KEY"] = os.getenv("GOOGLE_API_KEY")
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
```

### OPENAI_MODELS

```python
from langchain.chat_models import init_chat_model
models = init_chat_model("gpt-4.1")
models
```

```python
response = models.invoke("Hello, How are you?")
response.content
```

```python
from langchain_openai import ChatOpenAI
model = ChatOpenAI(model="gpt-4.1")
model.invoke("How are you?")
```

### GOOGLE_GEMINI_MODELS

```python
from langchain.chat_models import init_chat_model
models = init_chat_model("SPECIFIC-MODEL")
models

response = models.invoke("Hello, How are you?")
response.content
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

model = ChatGoogleGenerativeAI(model="specific-model")
model.invoke("How are you?")
```

### GROQ_MODELS

```python
from langchain.chat_models import init_chat_model

model = init_chat_model(model="llama-3.3-70b-versatile", model_provider="groq")

model.invoke("How are you?")
```

```python
from langchain_groq import ChatGroq

model = ChatGroq(model="llama-3.3-70b-versatile")
model.invoke("How are you?")
```

### Streaming and Batch

```python
model.invoke("How are you?")
```

### Streaming

```python
response = ""
for i in model.stream("How are you?"):  
    response += i.content + "-------------------"

print(response)
```

```python
# actual streaming (continuous stream)

for chunk in model.stream("Why do parrots have colorful feathers"):
    print(chunk.content, end="", flush=True)
```

### Batch

```python
#parallel processing

response = model.batch([
    "How are you",
    "How is your health"
])

for i in response:
    print(i.content)
```

```python
#config call

response = model.batch([
    "How are you",
    "How is your health",
    "How are you",
    "How is your health",
    "How are you",
    "How is your health"
],config = {
    'max_concurrency': 1
})

for i in response:
    print(i.content)
```
